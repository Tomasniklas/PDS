```python
import numpy as np
import matplotlib.pyplot as plt

# Frecuencia de muestreo: 2 MHz (2,000,000 muestras/segundo)
fs = 2e6

# Duración total: 1 milisegundo (suficiente para ver varios ciclos)

duracion = 1e-3   # 0.001 segundos

# Número total de muestras

N = int(fs * duracion)   # = 2000 muestras

# Vector de tiempo

t = np.linspace(0, duracion, N, endpoint=False)
```

 # Señal sinusoidal de 2 KHz #

```python
f1 = 2000     # frecuencia 2 kHz
A = 1         # amplitud

x1 = A * np.cos(2 * np.pi * f1 * t)
```

Es una señal de potencia debido a que es periodica y las señales periodicas tienen energía infinita dado que se repiten indefinidamente
En cambio, tiene potencia media finita, por lo que se clasifica como señal de potencia.
Período= 0.5 ms
Número de muestras: N=2000   (para 1 ms)

# Misma señal amplificada 3 dB y desfasada en π/2 #

```python
x2 = np.sqrt(2) * A * np.sin(2 * np.pi * f1 * t)
# también: np.sqrt(2) * A * np.cos(2 * np.pi * f1 * t - np.pi/2)
```

Señal de potencia tambien.

 # Señal modulada en amplitud (AM)

```python
fc = 1e6                       # 1 MHz
m = 1
x3 = (1 + m * x1) * np.cos(2 * np.pi * fc * t)
```

# Señal con saturación al 75% de su amplitud #

```python
x4 = np.clip(x1, -0.75, 0.75)
```

Sigue siendo periódica con período T =0.5 m

 # Señal cuadrada de 4 kHz

```python
f_cuad = 4000
x5 = np.sign(np.cos(2 * np.pi * f_cuad * t))
```

 # Pulso rectangular de 10 ms

```python
duracion_pulso = 15e-3        # 15 ms
N_pulso = int(fs * duracion_pulso)   # = 30000 muestras
t_pulso = np.linspace(0, duracion_pulso, N_pulso, endpoint=False)

pulso = np.zeros_like(t_pulso)
pulso[t_pulso < 10e-3] = 1    # amplitud 1 desde t=0 hasta t=10 ms
```
