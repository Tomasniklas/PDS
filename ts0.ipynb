```python
import numpy as np
import matplotlib.pyplot as plt

def funcion(vmax=1, dc=0, ff=1, ph=0, nn=1000, fs=1000):
    tt = np.arange(nn) / fs
    xx = dc + vmax * np.sin(2 * np.pi * ff * tt + ph)
    return tt.reshape(nn, 1), xx.reshape(nn, 1)

# Parámetros
N = 1000
fs = 1000

tt, xx = funcion(vmax=1, dc=0, ff=1, ph=0, nn=N, fs=fs)

print(tt.shape, xx.shape)

plt.plot(tt, xx)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.title('Señal senoidal')
plt.grid()
plt.show()

# Aliasing
frecuencias = [500, 999, 1001, 2001]

for f in frecuencias:
    tt, xx = funcion(vmax=1, dc=0, ff=f, ph=0, nn=N, fs=fs)
    plt.figure()
    plt.plot(tt, xx)
    plt.title(f'ff = {f} Hz')
    plt.grid()

plt.show()
```

```python
 Análisis

Se generó una señal senoidal 

Para frecuencias mayores a fs/2 = 500 Hz, se observa aliasing,
donde la señal aparenta tener una frecuencia menor debido al muestreo
```
